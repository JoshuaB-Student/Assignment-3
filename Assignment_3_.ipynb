```python
import random
```

```python
class ImageRegion:
  def __init__(self, x: int, y: int, width: int, height: int):
    self._x = x
    self._y = y
    self._width = width
    self._height = height
  def get_x(self) -> int:
        return self._x

  def get_y(self) -> int:
        return self._y

  def get_width(self) -> int:
        return self._width

  def get_height(self) -> int:
        return self._height

  def get_bounds(self) -> tuple:
        return (self._x, self._y, self._width, self._height)

  def get_center(self) -> tuple:
        cx = self._x + self._width // 2
        cy = self._y + self._height // 2
        return (cx, cy)
  def contains_point(self, px, py):
    return self._x <= px < self._x + self._width and self._y <= py < self._y + self._height
  def describe(self):
    return f"ImageRegion at ({self._x}, {self._y}), size {self._width}x{self.height}"
```

```python
class DifferenceRegion(ImageRegion):

  alteration_types = ["colour_shift", "blur", "brightness", "rotation",]

  def __init__(self, x, y, width, height, alteration_type = None):
    super(). __init__(x, y, width, height)
    self.alteration_type = alteration_type or random.choice(self.alteration_types)

    self._found = False

  def mark_as_found(self):
    self._found = True

  def is_found(self):
    return self._found

  def describe(self):
    status = "found" if self._found else "hidden"
    return f"DifferenceRegion at ({self._x}, {self._y})"
```
